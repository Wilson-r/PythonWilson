1. caricare i dati in un dataframe

```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/wtitze/4E/main/prime_titles.csv")
```

2. visualizzare i dati

```python
df
```

3. eliminare dal dataframe tutti i record con valori NaN

```python
df.dropna(subset= ["Name of the show"], inplace= True)
```

```python
df
```

4. calcolare il numero di show per ogni lingua. Ordinare in ordine decrescente sul numero

```python
dfLingue = df.groupby("Language").count().sort_values(by= ["Name of the show"], ascending= False)
dfLingue
```

5. visualizzare il grafico a barre verticali dei dati precedenti

```python
x = list(dfLingue.index)
y = list(dfLingue["Name of the show"])

fig,ax = plt.subplots()
fig.set_figwidth(15)
fig.set_figheight(10)
ax.set_title("FILM PER OGNI LINGUA", fontsize=20)
ax.set_xlabel("LINGUE", fontsize=15)
ax.set_ylabel("FILM", fontsize=15)
fig.autofmt_xdate(rotation=90)
ax.bar(x,y)
plt.savefig('foo2.png')
```

6. visualizzare le stesse informazioni dell'esercizio precedente utilizzando un grafico a torta

```python
labels = list(dfLingue.index)
sizes =list(dfLingue["Name of the show"])
fig1, ax1 = plt.subplots(figsize=(12, 12))
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.title("NUMERO DI RISTORANTI PER QUARTIERE", fontsize= 15 )
plt.show()
```
