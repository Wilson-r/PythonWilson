Visualizzare il totale di uomini vaccinati in ogni regione

```python
no2 = df.groupby("nome_area").sum()
no2["sesso_maschile"]
```

Visualizzare il totale di persone vaccinate in ogni regione

```python
no2["somma"]
```

Visualizzare la percentuale di persone vaccinate in ogni regione rispetto al totale dei vaccinati

```python
no2["somma"] / df["somma"].sum() * 100
```

Visualizzare le regioni in cui sono state vaccinate più di 1000000 di persone

```python
no3 = no2[no2["somma"] > 1000000]
no3
```

```python
labels = no3.index.values
sizes = no3['somma']
figura, x = plt.subplots(figsize=(10,10)) 
x.pie(sizes, labels = labels, autopct = '%1.1f%%', startangle = 90)
plt.show()
```

```python
labels = no3.index.values
values = no3['somma']
y = no3.plot.bar(figsize = (20, 10 ))
plt.bar(labels, values, color = ['red', 'blue','green'])
plt.title('vaccinazioni per regione')
y.get_legend().remove()
plt.show()
```

```python
labels = no3.index.values
values = no3['somma']
z = no3.plot.barh(x = labels, y = values)
plt.show()
```
