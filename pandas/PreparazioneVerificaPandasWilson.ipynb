```python
import pandas as pd
```

```python
df = pd.read_excel("https://github.com/wtitze/3E/blob/main/Medals.xlsx?raw=true")
df
```

eliminare colonna "Rank" "Total" "Rank by Total"

```python
#df = df.drop(columns= ["Rank", "Total", "Rank by Total"])
df.drop(columns=["Rank", "Total", "Rank by Total"], inplace= True)
df
```

cambiare nome colonna team/NOC in team

```python
df.rename(columns={"Team/NOC": "Team"}, inplace=True)
df
```

prendere colonna Totale della medaglie

```python
df["Total"] = df["Gold"] + df["Silver"]  + df["Bronze"]
df
```

nazioni con più di 80 medaglie


```python
df[df["Total"] > 80][["Team", "Total"]]
```

media delle medaglie

```python
df["Total"].mean()
```

nazioni con piu medaglie della media

```python
df[df["Total"] > df["Total"].mean()]
```
