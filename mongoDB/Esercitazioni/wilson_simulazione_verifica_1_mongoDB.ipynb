```python
#mongodb chiama i dizionari oggetti
# e le liste le chiama vettori
result=db.movies.find_one()
result
```

```python
print("trovare tutti i titoli dei film")
result=db.movies.find({},{"title": 1, "_id":0})
df=pd.DataFrame(list(result))
df
```

```python
print("I film che nel titolo hanno la parola the")
result=db.movies.aggregate([{"$match":{"title":{"$regex":"the"}}}])
df=pd.DataFrame(list(result))
df
```

```python
print("3.	I film americani degli anni 20")
result=db.movies.find({'$and': [{"countries" : "USA"} , {"year" : {"$gt" : 1919, "$lt" : 1930}}]},{"title": 1, "year" : 1, "countries": 1,"_id":0})
df=pd.DataFrame(list(result))
df
```

```python
print("4.	I film americani degli anni 30 e degli anni 40")
result=db.movies.find({'$and': [{"countries" : "USA"} , {"year" : {"$gt" : 1929, "$lt" : 1951}}]},{"title": 1, "year" : 1, "countries": 1,"_id":0})
df=pd.DataFrame(list(result))
df
```

```python
print("5.	Tutti i film tranne quelli americani")
result=db.movies.find({"countries" : {'$not': {"$eq":"USA"}}},{"title": 1, "year" : 1, "countries": 1,"_id":0})
df=pd.DataFrame(list(result))
df
```

```python
print("6.	I film che hanno un rating superiore a 4")
result=db.movies.find({"imdb.rating":{"$gt":4}},{"title": 1, "imdb" : 1,"_id":0})
df=pd.DataFrame(list(result))
df
```

```python
print("7.	Il numero di film per ogni lingua. Ordinare in ordine decrescente a partire della lingua più presente")
result=db.movies.aggregate( 
                            [{"$unwind": "$languages"},
                             {"$group": {"_id" : "$languages", "total" : {"$sum" : 1}}},
                             {"$sort": { "total": -1 }}]
)
df=pd.DataFrame(list(result))
df
```

```python

print("8. Le prime 3 lingue in cui sono stati girati i film")
result=db.movies.aggregate( 
                            [{"$unwind": "$languages"},
                             {"$group": {"_id" : "$languages", "total" : {"$sum" : 1}}},
                             {"$sort": { "total": -1 }},
                             {"$limit": 3}]
)
df=pd.DataFrame(list(result))
df
```

```python
print("9. Le lingue in cui sono stati girati meno di 50 film")

result=db.movies.aggregate( 
                            [{"$unwind": "$languages"},
                             {"$group": {"_id" : "$languages", "total" : {"$sum" : 1}}},
                             {"$sort": { "total": -1 }},
                             {"$match": {"total" : {"$lte" : 50}}}]
)
df=pd.DataFrame(list(result))
df
```

```python
print("10. La durata media dei film italiani")
result=db.movies.aggregate( 
                            [{"$unwind": "$languages"},
                             {"$match" : {"languages" : "Italian"}},
                             {"$group": {"_id": "null", "media": {"$avg":"$runtime"} } }                            
                            ]
)
df=pd.DataFrame(list(result))
df
```

```python
print("11. La durata media dei film delle varie nazioni")
result=db.movies.aggregate( 
                            [{"$unwind": "$countries"},
                             {"$group": {"_id": "$countries", "media": {"$avg":"$runtime"} } }                            
                            ]
)
df=pd.DataFrame(list(result))
df
```

```python
print("12. Il film che è dura di più")
result=db.movies.aggregate([{"$group" : {"_id" : "$runtime"}},
                            {"$sort": {"_id" : -1 }},
                            {"$limit" : 1}]
                           )
df=pd.DataFrame(list(result))
df
max_lenght = df.iloc[0,0]

result=db.movies.find({"runtime": max_lenght.astype(float)})
df=pd.DataFrame(list(result))
df
```
